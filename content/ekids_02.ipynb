# Вхідні дані програми(скрипта)

Джерелами вхідних даних є:

1. **Програміст**, той хто програму створив, може закласти в неї набір стандартних (`default`) вхідних даних.
2. **Користувач** програми, той хто виконує програму, може надати свої вхідні дані за допомогою клавіатури чи файлу.

2. В **файл** операційної системи `Linux`, `Windows` чи `MacOS`, мобільного пристрою
3. У БД (Базу Даних)
4. У відповідь на запит браузера користувача, тобто відобразиться на віддаленому пристрої.

# Перша функція

Ключове слово, що використовується для визначення функцій  - `def` від англійського **define** - визначати

`double` - довільне імʼя функції. Бажано щоб описувало основну роль, для чого використовується.

`x` - вхідні дані функції, аргумент

## Пробіли (ВІДСТУПИ) важливі !

Зверніть увагу, що внутрішній зміст функції знаходиться не по лівому краю, а дещо поодаль, **з відступом**

<hr />

При чому відступ визначає блок коду, який знаходиться всередині.

<hr />

А знак `:` (**двокрапка**) в кінці попереднього рядку позначає початок блоку коду з відступом.

````python
# plant_the_forest.py - програма, яка садить ліс Дубів

def plant_oak(name):
    message = "Ми посадили дуб на імʼя " + name
    return message


name_1 = "Велет"
name_2 = "Гулівер"
name_3 = "Титан"

# викличемо функцію так само як викликали `print()`
# за допомогою вказання круглих дужок

message_1 = plant_oak(name_1)
message_2 = plant_oak(name_2)
message_3 = plant_oak(name_3)

print(message_1, message_2, message_3, sep=" 🌱 ")
````

# Перша програма - Python скрипт

Тепер ви **вже** можете називатись програмістами, бо це реально працююча програма в мініатюрі.

Вона робить небагато, але в точності відповідає структурі того, як влаштовані всі на світі програми!

Цей скрипт містить **визначення функції**, якої не було в мові програмування.

```
def plant_oak(name):
    message = "Ми посадили дуб на імʼя " + name
    return message
```
цей блок коду
- починається із ключового слова `def` - це сигнал Пайтону що далі піде назва функції.
- після пробілу ми ввели довільне імʼя функції, в даному разі `plant_oak` (від англ. `plant` - садити)
- далі в дужках бачимо аргумент, або параметр з імʼям `name` - це теж довільне імʼя, його обрали ми.
- кінець рядку - завжди знак `:` - так Python розуміє, що далі буде блок коду **з відступом**
- всі подальші рядки коду будуть починатись з одного символу табуляції (⏭`TAB`) 
- `return` - ключове слово, яке сигналізує Python, що функція завершується та повертає вихідні дані в основну програму.

> Функції також називають ***підпрограми***

Коли ми почали наступний рядок **без відступу**

ми тим самим сигналізували Python про завершення внутрішнього блоку функції (який ще називають **тілом** функції)

Це не найліпший спосіб зберігання, але на разі цілком вистачить і цього.

> До наперед заданого рядку ми додали імʼя Дуба,
